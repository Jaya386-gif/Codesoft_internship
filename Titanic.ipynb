```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset (ensure the correct path)
data = pd.read_csv("Titanic-Dataset.csv")

# Select features and target variable, and make a copy to avoid warnings
x = data[['Pclass', 'Sex']].copy()
y = data['Survived']

# Convert categorical variable 'Sex' to numerical
x['Sex'] = x['Sex'].map({'male': 0, 'female': 1})

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
```

```python
len(x_train),len(x_test),len(y_test),len(y_train)
```

```python
Lreg=LogisticRegression(random_state=22)
Lreg.fit(x_train,y_train.ravel())
```

```python
y_predict=print(Lreg.predict(x_test))
y_predict
```
